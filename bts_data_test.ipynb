```python
import pandas as pd
import os
```

```python
bts_data_path = os.path.abspath("./bts_data")
sqlite_db_path = os.path.abspath("./bts_data/bts_data.db")

df_list = []

for filename in os.listdir(bts_data_path):
    if filename.endswith(".csv"):
        df = pd.read_csv(os.path.join(bts_data_path, filename))
        df_list.append(df)

df = pd.concat(df_list, ignore_index=True)

df[['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']] = df[['CARRIER_DELAY', 'WEATHER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']].fillna(0)
df
```

```python
df_bos = df[df['ORIGIN'] == 'BOS']
df_bos
```

```python
major_airports = ['ATL', 'AUS', 'BWI', 'CLT', 'MDW', 'ORD', 'DFW', 'DEN', 'DTW', 'FLL', 'IAH', 'LAS', 'LAX', 'MIA', 'MSP', 'BNA', 'JFK', 'LGA', 'EWR', 'MCO', 'PHL', 'PHX', 'SLC', 'SAN', 'SFO', 'SEA', 'TPA', 'DCA', 'IAD']
df_bos = df_bos[df_bos['DEST'].isin(major_airports)]

df_bos
```

```python
df_bos['MONTH_YEAR'] = pd.to_datetime(df_bos['FL_DATE']).dt.to_period('M')
df_bos
```

```python
df_majority = df_bos[df_bos['DEP_DEL15'] == 0]
df_minority = df_bos[df_bos['DEP_DEL15'] == 1]

n_samples_per_stratum = 100

def stratified_undersample(group, n_samples):
    if len(group) > n_samples:
        return group.sample(n=n_samples, random_state=42)
    else:
        return group
    
# Apply undersampling to the majority class while keeping the distribution
df_majority_undersampled = (
    df_majority.groupby('stratum', group_keys=False)
    .apply(lambda group: stratified_undersample(group, n_samples_per_stratum))
)

# Combine undersampled majority with all minority class rows
df_balanced = pd.concat([df_majority_undersampled, df_minority])
df_balanced = df_balanced.sample(frac=1).reset_index(drop=True)

df_balanced
```

```python
df_bos_sampled = df_balanced.sample(n=20000, random_state=24)

df_bos_sampled = df_bos_sampled.drop(columns=['MONTH_YEAR', 'stratum']).reset_index(drop=True)
df_bos_sampled
```
